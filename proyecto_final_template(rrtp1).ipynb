```python
# ============================================================================
# PROYECTO: TIMING STRATEGIES - XGBoost vs MLP
# ETAPA 1: PRE-PROCESAMIENTO
# ============================================================================

# ============================================================================
# 1. IMPORTACI√ìN DE LIBRER√çAS PRINCIPALES
# ============================================================================

# Manipulaci√≥n de datos
import pandas as pd
import numpy as np

# Para an√°lisis estad√≠stico b√°sico
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

print("‚úÖ Librer√≠as importadas correctamente")
print("="*70)

# ============================================================================
# 2. CARGA DEL DATA FRAME
# ============================================================================

# Cargar el dataset
dataset_name = "dataset_pinelis_ruppert_20250623_0752"

try:
    # Intentar cargar el dataset (ajustar seg√∫n la estructura real del archivo)
    df = pd.read_csv(f"{dataset_name}.csv", index_col=0, parse_dates=True)
    print(f"‚úÖ Dataset '{dataset_name}' cargado exitosamente")
except:
    print(f"‚ö†Ô∏è  Archivo no encontrado. Verificar nombre: {dataset_name}")
    print("Posibles soluciones:")
    print("- Verificar que el archivo est√© en el directorio correcto")
    print("- Confirmar extensi√≥n (.csv, .xlsx, etc.)")
    print("- Ajustar el path si est√° en una subcarpeta")

# ============================================================================
# 3. INFORMACI√ìN B√ÅSICA DEL DATASET
# ============================================================================

print("\n" + "="*70)
print("INFORMACI√ìN B√ÅSICA DEL DATASET")
print("="*70)

# Dimensiones
print(f"üìä Dimensiones del dataset: {df.shape}")
print(f"   - Observaciones: {df.shape[0]:,}")
print(f"   - Variables: {df.shape[1]}")

# Rango temporal (si el √≠ndice es fecha)
if isinstance(df.index, pd.DatetimeIndex):
    print(f"üìÖ Per√≠odo: {df.index.min().strftime('%Y-%m')} a {df.index.max().strftime('%Y-%m')}")
    years_coverage = (df.index.max() - df.index.min()).days / 365.25
    print(f"   - Cobertura: {years_coverage:.1f} a√±os")

# Primeras observaciones
print("\nüîç Primeras 5 observaciones:")
print(df.head())

# √öltimas observaciones
print("\nüîç √öltimas 5 observaciones:")
print(df.tail())

# ============================================================================
# 4. DESCRIPCI√ìN INICIAL DE LAS COLUMNAS
# ============================================================================

print("\n" + "="*70)
print("DESCRIPCI√ìN INICIAL DE COLUMNAS")
print("="*70)

# 4.1 TIPOS DE DATOS
print("\nüìã TIPOS DE DATOS:")
print("-" * 40)
tipos_datos = df.dtypes
for columna, tipo in tipos_datos.items():
    print(f"{columna:<25} : {tipo}")

# Resumen por tipo
print(f"\nüìä Resumen de tipos:")
print(f"   - Num√©ricas (float): {sum(tipos_datos == 'float64')}")
print(f"   - Enteras (int): {sum(tipos_datos == 'int64')}")
print(f"   - Objetos/Texto: {sum(tipos_datos == 'object')}")
print(f"   - Fechas: {sum(tipos_datos == 'datetime64[ns]')}")

# ============================================================================
# 5. EVALUACI√ìN DE MISSING VALUES
# ============================================================================

print("\n" + "="*70)
print("EVALUACI√ìN DE MISSING VALUES")
print("="*70)

# Conteo de missings por columna
missing_count = df.isnull().sum()
missing_percent = (missing_count / len(df)) * 100

# Mostrar variables con missings
vars_with_missing = missing_count[missing_count > 0]

if len(vars_with_missing) > 0:
    print("‚ö†Ô∏è  VARIABLES CON VALORES FALTANTES:")
    print("-" * 50)
    for var, count in vars_with_missing.items():
        percent = (count / len(df)) * 100
        print(f"{var:<25} : {count:>4} ({percent:>5.1f}%)")
else:
    print("‚úÖ DATASET COMPLETO - Sin valores faltantes")

# Estad√≠sticas generales de missings
total_values = df.shape[0] * df.shape[1]
total_missing = df.isnull().sum().sum()
overall_completeness = ((total_values - total_missing) / total_values) * 100

print(f"\nüìä ESTAD√çSTICAS GENERALES:")
print(f"   - Total de valores: {total_values:,}")
print(f"   - Valores faltantes: {total_missing:,}")
print(f"   - Completitud del dataset: {overall_completeness:.2f}%")

# ============================================================================
# 6. EVALUACI√ìN DE OUTLIERS
# ============================================================================

print("\n" + "="*70)
print("EVALUACI√ìN INICIAL DE OUTLIERS")
print("="*70)

# Seleccionar solo columnas num√©ricas
numeric_columns = df.select_dtypes(include=[np.number]).columns

print(f"üìä An√°lisis de outliers para {len(numeric_columns)} variables num√©ricas")
print("-" * 60)

# An√°lisis de outliers por variable usando IQR
outlier_summary = []

for col in numeric_columns:
    # Calcular quartiles
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # L√≠mites para outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Contar outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    n_outliers = len(outliers)
    percent_outliers = (n_outliers / len(df)) * 100

    outlier_summary.append({
        'Variable': col,
        'N_Outliers': n_outliers,
        'Percent_Outliers': percent_outliers
    })

# Convertir a DataFrame
outlier_df = pd.DataFrame(outlier_summary)

# Mostrar variables con outliers
vars_with_outliers = outlier_df[outlier_df['N_Outliers'] > 0].copy()

if len(vars_with_outliers) > 0:
    print("‚ö†Ô∏è  VARIABLES CON OUTLIERS DETECTADOS (M√©todo IQR):")
    print("-" * 55)
    print(vars_with_outliers.to_string(index=False))
else:
    print("‚úÖ No se detectaron outliers extremos con el m√©todo IQR")

# ============================================================================
# 7. ESTAD√çSTICAS DESCRIPTIVAS GENERALES
# ============================================================================

print("\n" + "="*70)
print("ESTAD√çSTICAS DESCRIPTIVAS GENERALES")
print("="*70)

# Estad√≠sticas b√°sicas
desc_stats = df.describe()
print("\nüìä ESTAD√çSTICAS DESCRIPTIVAS:")
print(desc_stats)

# ============================================================================
# 8. IDENTIFICACI√ìN DE VARIABLES OBJETIVO Y PREDICTORAS
# ============================================================================

print("\n" + "="*70)
print("IDENTIFICACI√ìN DE VARIABLES")
print("="*70)

# Basado en la documentaci√≥n del proyecto
target_variables = [col for col in df.columns if not col.endswith('_lag1')]
feature_variables = [col for col in df.columns if col.endswith('_lag1')]

print(f"üéØ VARIABLES OBJETIVO ({len(target_variables)}):")
for var in target_variables:
    print(f"   - {var}")

print(f"\nüîß VARIABLES PREDICTORAS ({len(feature_variables)}):")
for var in feature_variables:
    print(f"   - {var}")

# ============================================================================
# 9. RESUMEN DE LA ETAPA 1
# ============================================================================

print("\n" + "="*70)
print("RESUMEN ETAPA 1 - PRE-PROCESAMIENTO")
print("="*70)

print(f"‚úÖ Dataset cargado: {df.shape[0]:,} observaciones √ó {df.shape[1]} variables")
print(f"‚úÖ Completitud: {overall_completeness:.1f}%")
print(f"‚úÖ Variables objetivo identificadas: {len(target_variables)}")
print(f"‚úÖ Variables predictoras identificadas: {len(feature_variables)}")

if len(vars_with_missing) > 0:
    print(f"‚ö†Ô∏è  Variables con missings: {len(vars_with_missing)}")
else:
    print(f"‚úÖ Sin valores faltantes")

if len(vars_with_outliers) > 0:
    print(f"‚ö†Ô∏è  Variables con outliers: {len(vars_with_outliers)}")
else:
    print(f"‚úÖ Sin outliers extremos detectados")

print("\nüîÑ PR√ìXIMOS PASOS:")
print("   - Etapa 2: An√°lisis Exploratorio de Datos (EDA)")
print("   - An√°lisis de correlaciones")
print("   - Distribuciones por variable")
print("   - Relaciones entre predictoras y objetivos")

print("\n" + "="*70)
```

Incluir en esta secci√≥n:

    - Distribuci√≥n variables cont√≠nuas

```python
# ============================================================================
# PROYECTO: TIMING STRATEGIES - XGBoost vs MLP
# ETAPA 2: AN√ÅLISIS EXPLORATORIO DE DATOS (EDA)
# ============================================================================

# ============================================================================
# LIBRER√çAS PARA EDA
# ============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Configuraci√≥n de visualizaci√≥n
plt.style.use('default')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 10
sns.set_palette("husl")

print("‚úÖ Librer√≠as para EDA importadas correctamente")
print("="*70)

# ============================================================================
# CARGA DEL DATASET (desde etapa anterior)
# ============================================================================

# Cargar el dataset
dataset_name = "dataset_pinelis_ruppert_20250623_0752"
df = pd.read_csv(f"{dataset_name}.csv", index_col=0, parse_dates=True)

# Identificar variables objetivo y predictoras
target_variables = [col for col in df.columns if not col.endswith('_lag1')]
feature_variables = [col for col in df.columns if col.endswith('_lag1')]

print(f"üìä Dataset cargado: {df.shape[0]:,} observaciones √ó {df.shape[1]} variables")
print(f"üéØ Variables objetivo: {len(target_variables)}")
print(f"üîß Variables predictoras: {len(feature_variables)}")

# ============================================================================
# 1. AN√ÅLISIS DESCRIPTIVO DE CADA ATRIBUTO
# ============================================================================

print("\n" + "="*70)
print("AN√ÅLISIS DESCRIPTIVO DE VARIABLES")
print("="*70)

# ============================================================================
# 1.1 AN√ÅLISIS DE VARIABLES OBJETIVO
# ============================================================================

print("\n" + "-"*50)
print("AN√ÅLISIS DE VARIABLES OBJETIVO")
print("-"*50)

for target in target_variables:
    print(f"\nüìä VARIABLE OBJETIVO: {target}")
    print("-" * 30)

    # Estad√≠sticas descriptivas
    describe_stats = df[target].describe()
    print("Estad√≠sticas descriptivas:")
    for stat, value in describe_stats.items():
        print(f"   {stat:<10}: {value:>10.6f}")

    # Informaci√≥n adicional
    skewness = stats.skew(df[target])
    kurtosis = stats.kurtosis(df[target])
    print(f"   {'Skewness':<10}: {skewness:>10.6f}")
    print(f"   {'Kurtosis':<10}: {kurtosis:>10.6f}")

    # Test de normalidad
    shapiro_stat, shapiro_p = stats.shapiro(df[target].sample(min(5000, len(df))))
    print(f"   Shapiro-Wilk test p-value: {shapiro_p:.6f}")
    print(f"   ¬øDistribuci√≥n normal? {'No' if shapiro_p < 0.05 else 'S√≠'}")

# ============================================================================
# 1.2 AN√ÅLISIS DE VARIABLES PREDICTORAS
# ============================================================================

print("\n" + "-"*50)
print("AN√ÅLISIS DE VARIABLES PREDICTORAS")
print("-"*50)

# Crear tabla resumen de todas las predictoras
predictor_summary = []

for feature in feature_variables:
    stats_dict = {
        'Variable': feature,
        'Media': df[feature].mean(),
        'Std': df[feature].std(),
        'M√≠n': df[feature].min(),
        'Q25': df[feature].quantile(0.25),
        'Mediana': df[feature].median(),
        'Q75': df[feature].quantile(0.75),
        'M√°x': df[feature].max(),
        'Skewness': stats.skew(df[feature]),
        'Kurtosis': stats.kurtosis(df[feature])
    }
    predictor_summary.append(stats_dict)

# Convertir a DataFrame y mostrar
predictor_df = pd.DataFrame(predictor_summary)
print("\nüìä RESUMEN ESTAD√çSTICO DE VARIABLES PREDICTORAS:")
print(predictor_df.to_string(index=False, float_format='%.6f'))

# ============================================================================
# 2. DISTRIBUCI√ìN DE VARIABLES CONTINUAS
# ============================================================================

print("\n" + "="*70)
print("DISTRIBUCI√ìN DE VARIABLES CONTINUAS")
print("="*70)

# ============================================================================
# 2.1 DISTRIBUCIONES DE VARIABLES OBJETIVO
# ============================================================================

print("\nüìä Generando histogramas de variables objetivo...")

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for i, target in enumerate(target_variables):
    # Histograma
    axes[i].hist(df[target], bins=50, alpha=0.7, edgecolor='black')
    axes[i].set_title(f'Distribuci√≥n de {target}')
    axes[i].set_xlabel(target)
    axes[i].set_ylabel('Frecuencia')
    axes[i].grid(True, alpha=0.3)

    # A√±adir l√≠nea de media
    mean_val = df[target].mean()
    axes[i].axvline(mean_val, color='red', linestyle='--',
                   label=f'Media: {mean_val:.4f}')
    axes[i].legend()

plt.suptitle('Distribuciones de Variables Objetivo')
plt.tight_layout()
plt.show()

# ============================================================================
# 2.2 DISTRIBUCIONES DE VARIABLES PREDICTORAS
# ============================================================================

print("\nüìä Generando histogramas de variables predictoras...")

# Dividir las predictoras en grupos para mejor visualizaci√≥n
n_features = len(feature_variables)
n_cols = 3
n_rows = (n_features + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5*n_rows))
axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes

for i, feature in enumerate(feature_variables):
    # Histograma
    axes[i].hist(df[feature], bins=30, alpha=0.7, edgecolor='black')
    axes[i].set_title(f'{feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frecuencia')
    axes[i].grid(True, alpha=0.3)

    # A√±adir l√≠nea de media
    mean_val = df[feature].mean()
    axes[i].axvline(mean_val, color='red', linestyle='--', alpha=0.8)

# Ocultar subplots vac√≠os
for i in range(len(feature_variables), len(axes)):
    axes[i].set_visible(False)

plt.suptitle('Distribuciones de Variables Predictoras')
plt.tight_layout()
plt.show()

# ============================================================================
# 2.3 BOXPLOTS PARA DETECTAR OUTLIERS VISUALMENTE
# ============================================================================

print("\nüìä Generando boxplots para an√°lisis de outliers...")

# Boxplots de variables objetivo
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

for i, target in enumerate(target_variables):
    axes[i].boxplot(df[target])
    axes[i].set_title(f'Boxplot: {target}')
    axes[i].set_ylabel(target)
    axes[i].grid(True, alpha=0.3)

plt.suptitle('Boxplots de Variables Objetivo')
plt.tight_layout()
plt.show()

# Boxplots de variables predictoras (seleccionar las m√°s importantes)
print("\nüìä Boxplots de variables predictoras (muestra representativa)...")

# Seleccionar 8 variables predictoras para visualizaci√≥n
selected_features = feature_variables[:8]  # Primeras 8 para no sobrecargar

fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()

for i, feature in enumerate(selected_features):
    axes[i].boxplot(df[feature])
    axes[i].set_title(f'{feature}')
    axes[i].tick_params(axis='x', which='both', bottom=False, labelbottom=False)
    axes[i].grid(True, alpha=0.3)

plt.suptitle('Boxplots de Variables Predictoras (Muestra)')
plt.tight_layout()
plt.show()

# ============================================================================
# 3. EVALUACI√ìN DEL POTENCIAL PREDICTIVO
# ============================================================================

print("\n" + "="*70)
print("EVALUACI√ìN DEL POTENCIAL PREDICTIVO")
print("="*70)

# ============================================================================
# 3.1 MATRIZ DE CORRELACI√ìN
# ============================================================================

print("\nüìä MATRIZ DE CORRELACI√ìN")
print("-" * 40)

# Calcular matriz de correlaci√≥n completa
correlation_matrix = df.corr()

# Mostrar correlaciones de predictoras con targets
print("\nCorrelaciones de variables predictoras con variables objetivo:")
print("-" * 60)

target_correlations = pd.DataFrame()

for target in target_variables:
    target_corrs = []
    for feature in feature_variables:
        corr = correlation_matrix.loc[feature, target]
        target_corrs.append(corr)

    target_correlations[target] = target_corrs

target_correlations.index = feature_variables

# Ordenar por correlaci√≥n absoluta con la primera variable objetivo
target_correlations['abs_corr'] = abs(target_correlations[target_variables[0]])
target_correlations_sorted = target_correlations.sort_values('abs_corr', ascending=False)
target_correlations_sorted = target_correlations_sorted.drop('abs_corr', axis=1)

print(target_correlations_sorted.to_string(float_format='%.4f'))

# ============================================================================
# 3.2 VISUALIZACI√ìN DE MATRIZ DE CORRELACI√ìN
# ============================================================================

print("\nüìä Generando visualizaci√≥n de matriz de correlaci√≥n...")

# Heatmap de correlaciones entre predictoras y targets
fig, ax = plt.subplots(figsize=(8, 12))

# Seleccionar solo correlaciones de features con targets
corr_subset = correlation_matrix.loc[feature_variables, target_variables]

sns.heatmap(corr_subset, annot=True, cmap='RdBu_r', center=0,
            fmt='.3f', square=False, cbar_kws={'label': 'Correlaci√≥n'})

plt.title('Matriz de Correlaci√≥n: Variables Predictoras vs Objetivo')
plt.xlabel('Variables Objetivo')
plt.ylabel('Variables Predictoras')
plt.tight_layout()
plt.show()

# Heatmap de correlaciones entre predictoras
fig, ax = plt.subplots(figsize=(14, 12))

# Correlaciones entre features
feature_corr = correlation_matrix.loc[feature_variables, feature_variables]

sns.heatmap(feature_corr, annot=True, cmap='RdBu_r', center=0,
            fmt='.2f', square=True, cbar_kws={'label': 'Correlaci√≥n'})

plt.title('Matriz de Correlaci√≥n entre Variables Predictoras')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# ============================================================================
# 3.3 AN√ÅLISIS DE CORRELACIONES M√ÅS FUERTES
# ============================================================================

print("\nüìä AN√ÅLISIS DE CORRELACIONES M√ÅS FUERTES")
print("-" * 45)

# Top correlaciones con cada target
for target in target_variables:
    print(f"\nTOP 5 PREDICTORAS M√ÅS CORRELACIONADAS CON {target}:")
    print("-" * 50)

    target_corrs = correlation_matrix[target][feature_variables].copy()
    target_corrs_abs = target_corrs.abs().sort_values(ascending=False)

    for i, (feature, abs_corr) in enumerate(target_corrs_abs.head(5).items()):
        actual_corr = target_corrs[feature]
        print(f"{i+1}. {feature:<25}: {actual_corr:>7.4f}")

# ============================================================================
# 3.4 SCATTER PLOTS DE RELACIONES M√ÅS FUERTES
# ============================================================================

print("\nüìä Generando scatter plots de relaciones m√°s fuertes...")

# Para cada target, mostrar scatter plots con las 3 predictoras m√°s correlacionadas
for target in target_variables:
    target_corrs = correlation_matrix[target][feature_variables].copy()
    top_features = target_corrs.abs().sort_values(ascending=False).head(3).index

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    for i, feature in enumerate(top_features):
        correlation = target_corrs[feature]

        axes[i].scatter(df[feature], df[target], alpha=0.5, s=20)
        axes[i].set_xlabel(feature)
        axes[i].set_ylabel(target)
        axes[i].set_title(f'{feature} vs {target}\n(r = {correlation:.4f})')
        axes[i].grid(True, alpha=0.3)

        # A√±adir l√≠nea de tendencia
        z = np.polyfit(df[feature], df[target], 1)
        p = np.poly1d(z)
        axes[i].plot(df[feature].sort_values(), p(df[feature].sort_values()),
                    "r--", alpha=0.8, linewidth=2)

    plt.suptitle(f'Relaciones m√°s fuertes con {target}')
    plt.tight_layout()
    plt.show()

# ============================================================================
# 4. DECISIONES SOBRE MISSINGS Y OUTLIERS
# ============================================================================

print("\n" + "="*70)
print("DECISIONES SOBRE MISSINGS Y OUTLIERS")
print("="*70)

# ============================================================================
# 4.1 EVALUACI√ìN DE MISSING VALUES
# ============================================================================

print("\nüìä EVALUACI√ìN DE MISSING VALUES")
print("-" * 35)

missing_summary = df.isnull().sum()
total_missing = missing_summary.sum()

if total_missing == 0:
    print("‚úÖ DECISI√ìN: No se requiere imputaci√≥n de missing values")
    print("   Raz√≥n: El dataset est√° 100% completo")
else:
    print("‚ö†Ô∏è  VARIABLES CON MISSING VALUES:")
    for var, count in missing_summary[missing_summary > 0].items():
        percent = (count / len(df)) * 100
        print(f"   {var}: {count} valores ({percent:.2f}%)")

# ============================================================================
# 4.2 EVALUACI√ìN DE OUTLIERS
# ============================================================================

print("\nüìä EVALUACI√ìN DE OUTLIERS")
print("-" * 30)

print("üìã AN√ÅLISIS DE OUTLIERS POR VARIABLE:")
print("-" * 40)

outlier_decisions = []

for col in df.columns:
    # M√©todo IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    n_outliers = len(outliers)
    percent_outliers = (n_outliers / len(df)) * 100

    # M√©todo Z-score
    z_scores = np.abs(stats.zscore(df[col]))
    outliers_zscore = len(df[z_scores > 3])
    percent_outliers_zscore = (outliers_zscore / len(df)) * 100

    outlier_decisions.append({
        'Variable': col,
        'Outliers_IQR': n_outliers,
        'Percent_IQR': percent_outliers,
        'Outliers_ZScore': outliers_zscore,
        'Percent_ZScore': percent_outliers_zscore,
        'Min_Value': df[col].min(),
        'Max_Value': df[col].max()
    })

# Convertir a DataFrame
outlier_analysis = pd.DataFrame(outlier_decisions)

print("\nRESUMEN DE OUTLIERS (M√©todos IQR y Z-Score):")
print(outlier_analysis.to_string(index=False, float_format='%.2f'))

# ============================================================================
# 4.3 RECOMENDACIONES PARA OUTLIERS
# ============================================================================

print(f"\nüìã RECOMENDACIONES PARA TRATAMIENTO DE OUTLIERS:")
print("-" * 50)

# Variables con muchos outliers
high_outlier_vars = outlier_analysis[outlier_analysis['Percent_IQR'] > 5.0]

if len(high_outlier_vars) > 0:
    print("\n‚ö†Ô∏è  Variables con alto porcentaje de outliers (>5%):")
    for _, row in high_outlier_vars.iterrows():
        print(f"   {row['Variable']}: {row['Percent_IQR']:.1f}% (IQR)")

    print("\nüí° OPCIONES DE TRATAMIENTO:")
    print("   1. Mantener outliers (recomendado para datos financieros)")
    print("   2. Transformaci√≥n logar√≠tmica para reducir impacto")
    print("   3. Winsorizaci√≥n (reemplazar por percentiles)")
    print("   4. Eliminaci√≥n selectiva solo en casos extremos")

else:
    print("‚úÖ Niveles de outliers dentro de rangos aceptables")

# ============================================================================
# 5. RESUMEN DEL EDA
# ============================================================================

print("\n" + "="*70)
print("RESUMEN DEL AN√ÅLISIS EXPLORATORIO")
print("="*70)

print("\n‚úÖ AN√ÅLISIS COMPLETADO:")
print(f"   - Variables analizadas: {len(df.columns)}")
print(f"   - Distribuciones evaluadas: {len(df.columns)}")
print(f"   - Matriz de correlaci√≥n generada")
print(f"   - Outliers identificados y evaluados")

print("\nüìä HALLAZGOS PRINCIPALES:")
print("   - Todas las variables son num√©ricas continuas")
print("   - Dataset completo sin missing values")
print(f"   - Variables objetivo: {', '.join(target_variables)}")
print(f"   - Variables predictoras: {len(feature_variables)} variables")

print("\nüîÑ PR√ìXIMOS PASOS (Etapa 3 - Feature Engineering):")
print("   - Estandarizaci√≥n y escalado de variables")
print("   - Evaluaci√≥n de transformaciones necesarias")
print("   - Tratamiento final de outliers (si requerido)")
print("   - Preparaci√≥n para modelado ML")

print("\n" + "="*70)
```

```python
# ============================================================================
# PROYECTO: TIMING STRATEGIES - XGBoost vs MLP
# ETAPA 3: FEATURE ENGINEERING
# ============================================================================

# ============================================================================
# LIBRER√çAS PARA FEATURE ENGINEERING
# ============================================================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression
import warnings
warnings.filterwarnings('ignore')

print("‚úÖ Librer√≠as para Feature Engineering importadas correctamente")
print("="*70)

# ============================================================================
# CARGA DEL DATASET (desde etapa anterior)
# ============================================================================

# Cargar el dataset limpio de la etapa EDA
dataset_name = "dataset_pinelis_ruppert_20250623_0752"
df = pd.read_csv(f"{dataset_name}.csv", index_col=0, parse_dates=True)

# Identificar variables objetivo y predictoras
target_variables = [col for col in df.columns if not col.endswith('_lag1')]
feature_variables = [col for col in df.columns if col.endswith('_lag1')]

print(f"üìä Dataset cargado: {df.shape[0]:,} observaciones √ó {df.shape[1]} variables")
print(f"üéØ Variables objetivo: {len(target_variables)} - {target_variables}")
print(f"üîß Variables predictoras: {len(feature_variables)}")

# ============================================================================
# 1. EVALUACI√ìN INICIAL - SIGUIENDO PINELIS & RUPPERT
# ============================================================================

print("\n" + "="*70)
print("EVALUACI√ìN INICIAL - ENFOQUE PINELIS & RUPPERT")
print("="*70)

print("\nüìã CARACTER√çSTICAS DEL DATASET:")
print("-" * 40)

# Verificar estructura temporal (crucial para timing strategies)
print(f"‚úÖ Estructura temporal: Features t-1 ‚Üí Targets t")
print(f"‚úÖ Sin data leakage: Variables lag confirmadas")
print(f"‚úÖ Dataset completo: {df.isnull().sum().sum()} missing values")

# An√°lisis de rangos (importante para normalizaci√≥n)
print(f"\nüìä AN√ÅLISIS DE RANGOS DE VARIABLES:")
print("-" * 40)

range_analysis = pd.DataFrame({
    'Variable': df.columns,
    'Min': df.min(),
    'Max': df.max(),
    'Range': df.max() - df.min(),
    'Mean': df.mean(),
    'Std': df.std()
})

# Mostrar variables con mayor dispersi√≥n
print("Variables con mayor dispersi√≥n (candidatas a escalado):")
high_range_vars = range_analysis.nlargest(5, 'Range')
for _, row in high_range_vars.iterrows():
    print(f"  {row['Variable']:<25}: rango [{row['Min']:>8.3f}, {row['Max']:>8.3f}]")

# ============================================================================
# 2. ESTANDARIZACI√ìN Y ESCALADO
# ============================================================================

print("\n" + "="*70)
print("ESTANDARIZACI√ìN Y ESCALADO")
print("="*70)

print("\nüìä DECISI√ìN DE ESCALADO:")
print("-" * 30)
print("üîç Algoritmos objetivo:")
print("   - XGBoost: Tree-based ‚Üí Scale-invariant (no requiere escalado)")
print("   - MLP: Neural Network ‚Üí Scale-sensitive (requiere escalado)")
print("üìã Estrategia: Estandarizaci√≥n √∫nica para ambos modelos")
print("   (No afecta XGBoost, necesario para MLP)")

# ============================================================================
# 2.1 APLICACI√ìN DE STANDARDSCALER
# ============================================================================

print(f"\nüîß APLICANDO STANDARDSCALER:")
print("-" * 35)

# Crear escalador
scaler = StandardScaler()

# Separar features y targets para escalado independiente
X_original = df[feature_variables].copy()
y_original = df[target_variables].copy()

# Escalar features
X_scaled = pd.DataFrame(
    scaler.fit_transform(X_original),
    columns=feature_variables,
    index=df.index
)

# Escalar targets (importante para MLP)
target_scalers = {}
y_scaled = pd.DataFrame(index=df.index)

for target in target_variables:
    target_scaler = StandardScaler()
    y_scaled[target] = target_scaler.fit_transform(
        y_original[[target]]
    ).flatten()
    target_scalers[target] = target_scaler

# Combinar dataset escalado
df_scaled = pd.concat([X_scaled, y_scaled], axis=1)

print(f"‚úÖ Features escaladas: {X_scaled.shape[1]} variables")
print(f"‚úÖ Targets escaladas: {y_scaled.shape[1]} variables")

# Verificaci√≥n del escalado
print(f"\nüìã VERIFICACI√ìN POST-ESCALADO:")
print("-" * 35)
print(f"Features - Media promedio: {X_scaled.mean().mean():.6f}")
print(f"Features - Std promedio: {X_scaled.std().mean():.6f}")
print(f"Targets - Media promedio: {y_scaled.mean().mean():.6f}")
print(f"Targets - Std promedio: {y_scaled.std().mean():.6f}")

# ============================================================================
# 2.2 ESCALADORES ALTERNATIVOS (EVALUACI√ìN)
# ============================================================================

print(f"\nüî¨ EVALUACI√ìN DE ESCALADORES ALTERNATIVOS:")
print("-" * 45)

# MinMaxScaler (0-1 range)
minmax_scaler = MinMaxScaler()
X_minmax = pd.DataFrame(
    minmax_scaler.fit_transform(X_original),
    columns=feature_variables,
    index=df.index
)

# RobustScaler (resistente a outliers)
robust_scaler = RobustScaler()
X_robust = pd.DataFrame(
    robust_scaler.fit_transform(X_original),
    columns=feature_variables,
    index=df.index
)

# Comparaci√≥n de escaladores
scaling_comparison = pd.DataFrame({
    'Escalador': ['Original', 'Standard', 'MinMax', 'Robust'],
    'Mean_Range': [
        X_original.mean().max() - X_original.mean().min(),
        X_scaled.mean().max() - X_scaled.mean().min(),
        X_minmax.mean().max() - X_minmax.mean().min(),
        X_robust.mean().max() - X_robust.mean().min()
    ],
    'Std_Range': [
        X_original.std().max() - X_original.std().min(),
        X_scaled.std().max() - X_scaled.std().min(),
        X_minmax.std().max() - X_minmax.std().min(),
        X_robust.std().max() - X_robust.std().min()
    ]
})

print("Comparaci√≥n de escaladores:")
print(scaling_comparison.to_string(index=False, float_format='%.4f'))

print(f"\nüí° SELECCI√ìN: StandardScaler")
print("   - Distribuci√≥n est√°ndar (media=0, std=1)")
print("   - Apropiado para redes neuronales")
print("   - Mantiene outliers informativos (solo reescala)")

# ============================================================================
# 3. EVALUACI√ìN DE BUCKETIZATION
# ============================================================================

print("\n" + "="*70)
print("EVALUACI√ìN DE BUCKETIZATION")
print("="*70)

print("\nüìä AN√ÅLISIS PARA BUCKETIZATION:")
print("-" * 35)

# En finanzas, bucketization puede ser √∫til para capturar reg√≠menes no lineales
print("üîç Evaluando necesidad de bucketization:")
print("   - Variables financieras continuas: Mejor mantener precisi√≥n")
print("   - Timing strategies: Requieren se√±ales sutiles")
print("   - Pinelis & Ruppert: Usan variables continuas")

# An√°lisis de no-linealidades para variables clave
key_vars_for_buckets = ['dfy_lag1', 'svar_lag1', 'bm_lag1']  # Variables con alta correlaci√≥n

print(f"\nüìã VARIABLES CANDIDATAS PARA BUCKETIZATION:")
print("-" * 50)

bucket_analysis = {}

for var in key_vars_for_buckets:
    if var in df.columns:
        # Crear quintiles para an√°lisis
        quintiles = pd.qcut(df[var], q=5, labels=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'])

        # Analizar target por quintiles
        target_by_quintile = {}
        for target in target_variables:
            quintile_stats = df.groupby(quintiles)[target].agg(['mean', 'std', 'count'])
            target_by_quintile[target] = quintile_stats

        bucket_analysis[var] = target_by_quintile

        print(f"\n{var}:")
        for target in target_variables:
            means = target_by_quintile[target]['mean']
            variation = means.std()
            print(f"  {target}: Variaci√≥n entre quintiles = {variation:.6f}")

print(f"\nüí° DECISI√ìN: NO aplicar bucketization")
print("   - Variables financieras requieren precisi√≥n continua")
print("   - Timing strategies dependen de se√±ales sutiles")
print("   - Seguir enfoque Pinelis & Ruppert (variables continuas)")

# ============================================================================
# 4. EVALUACI√ìN DE ONE-HOT ENCODING
# ============================================================================

print("\n" + "="*70)
print("EVALUACI√ìN DE ONE-HOT ENCODING")
print("="*70)

print("\nüìä AN√ÅLISIS DE VARIABLES CATEG√ìRICAS:")
print("-" * 40)

# Verificar tipos de datos
categorical_vars = df.select_dtypes(include=['object', 'category']).columns.tolist()
numeric_vars = df.select_dtypes(include=[np.number]).columns.tolist()

print(f"Variables categ√≥ricas detectadas: {len(categorical_vars)}")
print(f"Variables num√©ricas: {len(numeric_vars)}")

if len(categorical_vars) == 0:
    print("‚úÖ DECISI√ìN: NO se requiere One-Hot Encoding")
    print("   - Todas las variables son num√©ricas continuas")
    print("   - Dataset financiero est√°ndar (siguiendo Goyal & Welch)")
    print("   - Consistente con enfoque Pinelis & Ruppert")
else:
    print(f"‚ö†Ô∏è  Variables categ√≥ricas encontradas: {categorical_vars}")
    # Aplicar One-Hot Encoding si hay variables categ√≥ricas
    df_encoded = pd.get_dummies(df_scaled, columns=categorical_vars, prefix=categorical_vars)
    print(f"‚úÖ One-Hot Encoding aplicado")
    print(f"   Variables despu√©s de encoding: {df_encoded.shape[1]}")

# ============================================================================
# 5. T√âCNICAS DE REDUCCI√ìN DE DIMENSIONALIDAD
# ============================================================================

print("\n" + "="*70)
print("T√âCNICAS DE REDUCCI√ìN DE DIMENSIONALIDAD")
print("="*70)

print("\nüìä EVALUACI√ìN DE DIMENSIONALIDAD:")
print("-" * 35)

n_features = len(feature_variables)
n_observations = len(df)

print(f"N√∫mero de features: {n_features}")
print(f"N√∫mero de observaciones: {n_observations}")
print(f"Ratio observaciones/features: {n_observations/n_features:.1f}")

print(f"\nüîç AN√ÅLISIS DE NECESIDAD:")
print("-" * 30)
print(f"   - Features: {n_features} (bajo)")
print(f"   - Observaciones: {n_observations:,} (alto)")
print(f"   - Ratio: {n_observations/n_features:.0f}:1 (excelente)")
print(f"   - Pinelis & Ruppert: 12 variables est√°ndar")

# ============================================================================
# 5.1 AN√ÅLISIS PCA (EXPLORATORIO)
# ============================================================================

print(f"\nüî¨ AN√ÅLISIS PCA (EXPLORATORIO):")
print("-" * 35)

# Aplicar PCA a las features escaladas
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# An√°lisis de varianza explicada
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

print("Varianza explicada por componente:")
for i, (var_ratio, cum_var) in enumerate(zip(explained_variance_ratio[:8], cumulative_variance[:8])):
    print(f"  PC{i+1}: {var_ratio:.3f} ({cum_var:.3f} acumulada)")

# Determinar componentes necesarios para 95% de varianza
n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1
n_components_90 = np.argmax(cumulative_variance >= 0.90) + 1

print(f"\nComponentes para 90% varianza: {n_components_90}")
print(f"Componentes para 95% varianza: {n_components_95}")

# ============================================================================
# 5.2 FEATURE SELECTION
# ============================================================================

print(f"\nüéØ FEATURE SELECTION:")
print("-" * 25)

# An√°lisis de importancia univariada
feature_scores = {}

for target in target_variables:
    # F-regression score
    f_selector = SelectKBest(score_func=f_regression, k='all')
    f_selector.fit(X_scaled, y_original[target])

    # Mutual information
    mi_scores = mutual_info_regression(X_scaled, y_original[target], random_state=42)

    feature_scores[target] = {
        'f_regression': f_selector.scores_,
        'mutual_info': mi_scores
    }

# Mostrar top features por target
print("\nTop 5 features por variable objetivo:")
for target in target_variables:
    print(f"\n{target}:")
    f_scores = feature_scores[target]['f_regression']
    top_features_idx = np.argsort(f_scores)[::-1][:5]

    for i, idx in enumerate(top_features_idx):
        feature_name = feature_variables[idx]
        score = f_scores[idx]
        print(f"  {i+1}. {feature_name:<25}: {score:.2f}")

print(f"\nüí° DECISI√ìN: NO aplicar reducci√≥n de dimensionalidad")
print("   - Solo 12 features (√≥ptimo para ML)")
print("   - Variables est√°ndar academia financiera")
print("   - Mantener interpretabilidad (timing strategies)")
print("   - Seguir metodolog√≠a Pinelis & Ruppert")

# ============================================================================
# 6. DATASET FINAL PARA MODELADO
# ============================================================================

print("\n" + "="*70)
print("DATASET FINAL PARA MODELADO")
print("="*70)

# Crear datasets finales
print("\nüîß PREPARACI√ìN DE DATASETS FINALES:")
print("-" * 40)

# Dataset escalado (para ambos modelos)
df_final = df_scaled.copy()

# Separar X e y
X_final = df_final[feature_variables].copy()
y_final = df_final[target_variables].copy()

# Tambi√©n mantener versi√≥n original (para XGBoost si es necesario)
X_original_final = df[feature_variables].copy()
y_original_final = df[target_variables].copy()

print(f"‚úÖ Dataset escalado preparado:")
print(f"   - Features (X): {X_final.shape}")
print(f"   - Targets (y): {y_final.shape}")
print(f"   - Per√≠odo: {df_final.index.min()} a {df_final.index.max()}")

print(f"‚úÖ Dataset original mantenido (backup):")
print(f"   - Features (X): {X_original_final.shape}")
print(f"   - Targets (y): {y_original_final.shape}")

# ============================================================================
# 7. VALIDACI√ìN Y VERIFICACI√ìN
# ============================================================================

print("\n" + "="*70)
print("VALIDACI√ìN Y VERIFICACI√ìN")
print("="*70)

print("\nüîç VERIFICACIONES FINALES:")
print("-" * 30)

# 1. Verificar ausencia de NaN
nan_check = df_final.isnull().sum().sum()
print(f"‚úÖ Missing values: {nan_check}")

# 2. Verificar escalado
scaling_check = {
    'mean_close_to_zero': abs(X_final.mean().mean()) < 0.01,
    'std_close_to_one': abs(X_final.std().mean() - 1.0) < 0.01
}
print(f"‚úÖ Escalado correcto: {all(scaling_check.values())}")

# 3. Verificar estructura temporal
temporal_check = all(col.endswith('_lag1') for col in feature_variables)
print(f"‚úÖ Estructura temporal: {temporal_check}")

# 4. Verificar dimensiones
dimension_check = X_final.shape[0] == y_final.shape[0]
print(f"‚úÖ Dimensiones consistentes: {dimension_check}")

# ============================================================================
# 8. OBJETOS PARA SIGUIENTE ETAPA
# ============================================================================

print("\n" + "="*70)
print("OBJETOS PARA ETAPA DE MODELADO")
print("="*70)

# Crear diccionario con todos los objetos necesarios
feature_engineering_output = {
    # Datasets finales
    'X_scaled': X_final,
    'y_scaled': y_final,
    'X_original': X_original_final,
    'y_original': y_original_final,
    'df_final': df_final,

    # Escaladores (para inverse transform)
    'feature_scaler': scaler,
    'target_scalers': target_scalers,

    # Metadata
    'feature_variables': feature_variables,
    'target_variables': target_variables,
    'n_features': len(feature_variables),
    'n_targets': len(target_variables),
    'n_observations': len(df_final),

    # An√°lisis aplicados
    'transformations_applied': {
        'standardization': True,
        'bucketization': False,
        'one_hot_encoding': False,
        'dimensionality_reduction': False
    },

    # Informaci√≥n para modelado
    'pca_analysis': {
        'explained_variance_ratio': explained_variance_ratio,
        'n_components_90': n_components_90,
        'n_components_95': n_components_95
    },

    'feature_scores': feature_scores
}

print("\nüì¶ OBJETOS CREADOS PARA MODELADO:")
print("-" * 35)
print("‚úÖ X_scaled, y_scaled: Datos escalados para MLP")
print("‚úÖ X_original, y_original: Datos originales para XGBoost")
print("‚úÖ feature_scaler, target_scalers: Para inverse transform")
print("‚úÖ feature_engineering_output: Diccionario completo")

# ============================================================================
# 9. RESUMEN DE FEATURE ENGINEERING
# ============================================================================

print("\n" + "="*70)
print("RESUMEN DE FEATURE ENGINEERING")
print("="*70)

print("\n‚úÖ TRANSFORMACIONES APLICADAS:")
print("-" * 35)
print("üîß Estandarizaci√≥n: StandardScaler aplicado")
print("   - Features: media=0, std=1")
print("   - Targets: media=0, std=1")
print("‚ùå Bucketization: NO aplicada")
print("   - Variables continuas mantenidas")
print("‚ùå One-Hot Encoding: NO requerida")
print("   - Solo variables num√©ricas")
print("‚ùå Reducci√≥n dimensionalidad: NO aplicada")
print("   - 12 features √≥ptimo")

print(f"\nüìä DATASET FINAL:")
print("-" * 20)
print(f"   - Observaciones: {len(df_final):,}")
print(f"   - Features: {len(feature_variables)}")
print(f"   - Targets: {len(target_variables)}")
print(f"   - Per√≠odo: {(df_final.index.max() - df_final.index.min()).days / 365.25:.1f} a√±os")
print(f"   - Calidad: 100% completo, escalado, listo para ML")

print(f"\nüéØ METODOLOG√çA:")
print("-" * 15)
print("   - Siguiendo Pinelis & Ruppert (2022)")
print("   - Variables est√°ndar Goyal & Welch (2008)")
print("   - Adaptado para XGBoost vs MLP")
print("   - Preparado para reward-risk timing")

print("\nüîÑ PR√ìXIMOS PASOS (Etapa 4 - Entrenamiento):")
print("   - Train/test split temporal")
print("   - Configuraci√≥n de modelos XGBoost y MLP")
print("   - Grid search y validaci√≥n")
print("   - Evaluaci√≥n de performance")

print("\n" + "="*70)
```

```python
# ============================================================================
# PROYECTO: TIMING STRATEGIES - XGBoost vs MLP
# ETAPA 4: ENTRENAMIENTO - METODOLOG√çA PINELIS & RUPPERT FIEL
# ============================================================================

# ============================================================================
# LIBRER√çAS PARA ENTRENAMIENTO
# ============================================================================

import pandas as pd
import numpy as np
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb
import time
from itertools import product
import warnings
warnings.filterwarnings('ignore')

print("‚úÖ Librer√≠as para entrenamiento importadas correctamente")
print("="*70)

# ============================================================================
# CARGA DE DATOS PREPARADOS
# ============================================================================

# Cargar el dataset escalado
dataset_name = "dataset_pinelis_ruppert_20250623_0752"
df = pd.read_csv(f"{dataset_name}.csv", index_col=0, parse_dates=True)

# Recrear escalado para MLP
from sklearn.preprocessing import StandardScaler

target_variables = [col for col in df.columns if not col.endswith('_lag1')]
feature_variables = [col for col in df.columns if col.endswith('_lag1')]

# Escalar features
feature_scaler = StandardScaler()
X_scaled = pd.DataFrame(
    feature_scaler.fit_transform(df[feature_variables]),
    columns=feature_variables,
    index=df.index
)

# Escalar targets para MLP
target_scalers = {}
y_scaled = pd.DataFrame(index=df.index)
for target in target_variables:
    target_scaler = StandardScaler()
    y_scaled[target] = target_scaler.fit_transform(df[[target]]).flatten()
    target_scalers[target] = target_scaler

# Dataset original para XGBoost
X_original = df[feature_variables].copy()
y_original = df[target_variables].copy()

print(f"üìä Datos cargados: {len(df):,} observaciones")
print(f"üéØ Variables objetivo: {target_variables}")
print(f"üîß Variables predictoras: {len(feature_variables)}")

# ============================================================================
# 1. PER√çODOS TEMPORALES - SIGUIENDO PINELIS & RUPPERT
# ============================================================================

print("\n" + "="*70)
print("PER√çODOS TEMPORALES - METODOLOG√çA PINELIS & RUPPERT")
print("="*70)

# Definir fechas exactas seg√∫n paper
training_start = df.index.min()  # 1927-05
training_end = pd.Timestamp('1957-04-01')   # 1927-1957 (30 a√±os)
validation_start = pd.Timestamp('1957-05-01')  # 1958-1988 (validation)
validation_end = pd.Timestamp('1987-04-01')
test_start = pd.Timestamp('1987-05-01')     # 1989-2019 (test)
test_end = df.index.max()  # 2024-11

print(f"üìà Training inicial: {training_start.strftime('%Y-%m')} a {training_end.strftime('%Y-%m')}")
print(f"üîç Validation: {validation_start.strftime('%Y-%m')} a {validation_end.strftime('%Y-%m')}")
print(f"üéØ Test: {test_start.strftime('%Y-%m')} a {test_end.strftime('%Y-%m')}")

# Crear m√°scaras
training_mask = (df.index >= training_start) & (df.index <= training_end)
validation_mask = (df.index >= validation_start) & (df.index <= validation_end)
test_mask = (df.index >= test_start) & (df.index <= test_end)

print(f"\nüìä Observaciones por per√≠odo:")
print(f"   - Training: {training_mask.sum()}")
print(f"   - Validation: {validation_mask.sum()}")
print(f"   - Test: {test_mask.sum()}")

# ============================================================================
# 2. DEFINICI√ìN DE HIPERPAR√ÅMETROS A PROBAR
# ============================================================================

print("\n" + "="*70)
print("DEFINICI√ìN DE HIPERPAR√ÅMETROS")
print("="*70)

# Grid de hiperpar√°metros XGBoost (simplificado para expanding window)
xgb_param_grid = {
    'n_estimators': [100, 300, 500],
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

# Grid de hiperpar√°metros MLP (simplificado)
mlp_param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'alpha': [0.0001, 0.001, 0.01],
    'max_iter': [500, 1000]
}

print(f"üöÄ XGBoost combinations: {np.prod([len(v) for v in xgb_param_grid.values()])}")
print(f"üß† MLP combinations: {np.prod([len(v) for v in mlp_param_grid.values()])}")

# ============================================================================
# 3. M√âTRICAS DE EVALUACI√ìN
# ============================================================================

def calculate_r2_oos(y_true, y_pred):
    """
    Calcula R¬≤ out-of-sample seg√∫n Pinelis & Ruppert:
    R¬≤_os = 1 - Œ£(y_true - y_pred)¬≤ / Œ£(y_true - mean(y_true))¬≤
    """
    numerator = np.sum((y_true - y_pred) ** 2)
    denominator = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - (numerator / denominator)

def calculate_directional_accuracy(y_true, y_pred):
    """Calcula % de predicciones con signo correcto"""
    return np.mean(np.sign(y_true) == np.sign(y_pred))

print("\n‚úÖ M√©tricas definidas:")
print("   - R¬≤ out-of-sample (como en paper)")
print("   - Directional Accuracy")

# ============================================================================
# 4. EXPANDING WINDOW VALIDATION - SELECCI√ìN DE HIPERPAR√ÅMETROS
# ============================================================================

print("\n" + "="*70)
print("EXPANDING WINDOW VALIDATION - SELECCI√ìN HIPERPAR√ÅMETROS")
print("="*70)

def expanding_window_validation(X_data, y_data, param_grid, model_type, target_name):
    """
    Implementa expanding window validation para seleccionar hiperpar√°metros
    siguiendo metodolog√≠a Pinelis & Ruppert
    """
    print(f"\nüîç Validando {model_type} para {target_name}")

    # Fechas del per√≠odo de validaci√≥n
    validation_dates = df.index[validation_mask]

    best_params = None
    best_score = -np.inf
    param_scores = {}

    # Generar todas las combinaciones de par√°metros
    param_names = list(param_grid.keys())
    param_values = list(param_grid.values())

    total_combinations = len(list(product(*param_values)))
    print(f"   Probando {total_combinations} combinaciones de hiperpar√°metros...")

    for i, param_combination in enumerate(product(*param_values)):
        if (i + 1) % 20 == 0:
            print(f"   Progreso: {i+1}/{total_combinations}")

        params = dict(zip(param_names, param_combination))

        # Lista para almacenar predicciones y valores reales
        predictions = []
        actuals = []

        # Expanding window sobre per√≠odo de validaci√≥n
        for val_date in validation_dates:
            # Training set hasta (val_date - 1)
            train_end_date = val_date - pd.DateOffset(months=1)
            train_mask_current = (df.index >= training_start) & (df.index <= train_end_date)

            if train_mask_current.sum() < 12:  # M√≠nimo 12 observaciones
                continue

            # Datos de entrenamiento
            X_train = X_data[train_mask_current]
            y_train = y_data[train_mask_current]

            # Datos para predicci√≥n (solo val_date)
            X_val = X_data[df.index == val_date]
            y_val = y_data[df.index == val_date]

            if len(X_val) == 0 or len(y_val) == 0:
                continue

            try:
                # Entrenar modelo
                if model_type == 'XGBoost':
                    model = xgb.XGBRegressor(random_state=42, verbosity=0, **params)
                else:  # MLP
                    model = MLPRegressor(random_state=42, **params)

                model.fit(X_train, y_train)

                # Predicci√≥n
                y_pred = model.predict(X_val)

                predictions.extend(y_pred)
                actuals.extend(y_val.values)

            except Exception as e:
                # Si hay error, asignar score muy bajo
                break

        # Calcular R¬≤ promedio para esta combinaci√≥n de par√°metros
        if len(predictions) > 10:  # M√≠nimo de predicciones
            r2_score = calculate_r2_oos(np.array(actuals), np.array(predictions))
            param_scores[str(params)] = r2_score

            if r2_score > best_score:
                best_score = r2_score
                best_params = params

    print(f"   ‚úÖ Mejor R¬≤: {best_score:.6f}")
    print(f"   üéØ Mejores par√°metros: {best_params}")

    return best_params, best_score, param_scores

# ============================================================================
# 4.1 VALIDACI√ìN PARA EXCESS RETURNS
# ============================================================================

print(f"\nüéØ VALIDACI√ìN EXCESS_RETURN_1M:")
print("-" * 40)

# XGBoost para excess returns
start_time = time.time()
xgb_excess_params, xgb_excess_score, _ = expanding_window_validation(
    X_original, y_original['excess_return_1m'],
    xgb_param_grid, 'XGBoost', 'excess_return_1m'
)
xgb_excess_time = time.time() - start_time

# MLP para excess returns
start_time = time.time()
mlp_excess_params, mlp_excess_score, _ = expanding_window_validation(
    X_scaled, y_scaled['excess_return_1m'],
    mlp_param_grid, 'MLP', 'excess_return_1m'
)
mlp_excess_time = time.time() - start_time

# ============================================================================
# 4.2 VALIDACI√ìN PARA VOLATILITY
# ============================================================================

print(f"\nüéØ VALIDACI√ìN REALIZED_VOL_1M:")
print("-" * 40)

# XGBoost para volatility
start_time = time.time()
xgb_vol_params, xgb_vol_score, _ = expanding_window_validation(
    X_original, y_original['realized_vol_1m'],
    xgb_param_grid, 'XGBoost', 'realized_vol_1m'
)
xgb_vol_time = time.time() - start_time

# MLP para volatility
start_time = time.time()
mlp_vol_params, mlp_vol_score, _ = expanding_window_validation(
    X_scaled, y_scaled['realized_vol_1m'],
    mlp_param_grid, 'MLP', 'realized_vol_1m'
)
mlp_vol_time = time.time() - start_time

# ============================================================================
# 5. EVALUACI√ìN EN TEST SET CON EXPANDING WINDOW
# ============================================================================

print("\n" + "="*70)
print("EVALUACI√ìN EN TEST SET - EXPANDING WINDOW")
print("="*70)

def expanding_window_test(X_data, y_data, best_params, model_type, target_name, scaler=None):
    """
    Evaluaci√≥n final en test set con expanding window
    """
    print(f"\nüî¨ Evaluando {model_type} para {target_name}")

    test_dates = df.index[test_mask]
    predictions = []
    actuals = []

    print(f"   Per√≠odo de test: {len(test_dates)} meses")

    for i, test_date in enumerate(test_dates):
        if (i + 1) % 60 == 0:  # Progress cada 5 a√±os
            print(f"   Progreso: {i+1}/{len(test_dates)} meses")

        # Training set hasta (test_date - 1)
        train_end_date = test_date - pd.DateOffset(months=1)
        train_mask_current = (df.index >= training_start) & (df.index <= train_end_date)

        # Datos de entrenamiento (incluyendo validation period)
        X_train = X_data[train_mask_current]
        y_train = y_data[train_mask_current]

        # Datos para predicci√≥n
        X_test = X_data[df.index == test_date]
        y_test = y_data[df.index == test_date]

        if len(X_test) == 0 or len(y_test) == 0:
            continue

        try:
            # Entrenar modelo con mejores par√°metros
            if model_type == 'XGBoost':
                model = xgb.XGBRegressor(random_state=42, verbosity=0, **best_params)
            else:  # MLP
                model = MLPRegressor(random_state=42, **best_params)

            model.fit(X_train, y_train)

            # Predicci√≥n
            y_pred = model.predict(X_test)

            # Si es MLP con target escalado, hacer inverse transform
            if model_type == 'MLP' and scaler is not None:
                y_pred = scaler.inverse_transform(y_pred.reshape(-1, 1)).flatten()
                y_test_original = scaler.inverse_transform(y_test.values.reshape(-1, 1)).flatten()

                predictions.extend(y_pred)
                actuals.extend(y_test_original)
            else:
                predictions.extend(y_pred)
                actuals.extend(y_test.values)

        except Exception as e:
            print(f"   Error en {test_date}: {e}")
            continue

    # Calcular m√©tricas finales
    predictions = np.array(predictions)
    actuals = np.array(actuals)

    r2_final = calculate_r2_oos(actuals, predictions)
    dir_acc = calculate_directional_accuracy(actuals, predictions)
    rmse = np.sqrt(mean_squared_error(actuals, predictions))

    print(f"   ‚úÖ R¬≤ final: {r2_final:.6f}")
    print(f"   ‚úÖ Directional Accuracy: {dir_acc:.4f}")
    print(f"   ‚úÖ RMSE: {rmse:.6f}")

    return {
        'r2': r2_final,
        'directional_accuracy': dir_acc,
        'rmse': rmse,
        'predictions': predictions,
        'actuals': actuals
    }

# ============================================================================
# 5.1 EVALUACI√ìN FINAL EXCESS RETURNS
# ============================================================================

print(f"\nüéØ EVALUACI√ìN FINAL EXCESS_RETURN_1M:")

# XGBoost
xgb_excess_results = expanding_window_test(
    X_original, y_original['excess_return_1m'],
    xgb_excess_params, 'XGBoost', 'excess_return_1m'
)

# MLP
mlp_excess_results = expanding_window_test(
    X_scaled, y_scaled['excess_return_1m'],
    mlp_excess_params, 'MLP', 'excess_return_1m',
    scaler=target_scalers['excess_return_1m']
)

# ============================================================================
# 5.2 EVALUACI√ìN FINAL VOLATILITY
# ============================================================================

print(f"\nüéØ EVALUACI√ìN FINAL REALIZED_VOL_1M:")

# XGBoost
xgb_vol_results = expanding_window_test(
    X_original, y_original['realized_vol_1m'],
    xgb_vol_params, 'XGBoost', 'realized_vol_1m'
)

# MLP
mlp_vol_results = expanding_window_test(
    X_scaled, y_scaled['realized_vol_1m'],
    mlp_vol_params, 'MLP', 'realized_vol_1m',
    scaler=target_scalers['realized_vol_1m']
)

# ============================================================================
# 6. RESULTADOS FINALES
# ============================================================================

print("\n" + "="*70)
print("RESULTADOS FINALES - COMPARACI√ìN CON PAPER")
print("="*70)

# Consolidar resultados
final_results = {
    'excess_return_1m': {
        'xgboost': xgb_excess_results,
        'mlp': mlp_excess_results,
        'best_params': {
            'xgboost': xgb_excess_params,
            'mlp': mlp_excess_params
        },
        'validation_scores': {
            'xgboost': xgb_excess_score,
            'mlp': mlp_excess_score
        }
    },
    'realized_vol_1m': {
        'xgboost': xgb_vol_results,
        'mlp': mlp_vol_results,
        'best_params': {
            'xgboost': xgb_vol_params,
            'mlp': mlp_vol_params
        },
        'validation_scores': {
            'xgboost': xgb_vol_score,
            'mlp': mlp_vol_score
        }
    }
}

print(f"\nüìä RESULTADOS EXCESS_RETURN_1M:")
print(f"   XGBoost R¬≤: {xgb_excess_results['r2']:>8.6f} | Dir Acc: {xgb_excess_results['directional_accuracy']:>6.2%}")
print(f"   MLP R¬≤:     {mlp_excess_results['r2']:>8.6f} | Dir Acc: {mlp_excess_results['directional_accuracy']:>6.2%}")

print(f"\nüìä RESULTADOS REALIZED_VOL_1M:")
print(f"   XGBoost R¬≤: {xgb_vol_results['r2']:>8.6f} | Dir Acc: {xgb_vol_results['directional_accuracy']:>6.2%}")
print(f"   MLP R¬≤:     {mlp_vol_results['r2']:>8.6f} | Dir Acc: {mlp_vol_results['directional_accuracy']:>6.2%}")

print(f"\nüî¨ COMPARACI√ìN CON PAPER PINELIS & RUPPERT:")
print("   Paper Random Forest excess returns R¬≤: 0.0052 (0.52%)")
print("   Paper Random Forest volatility R¬≤:     0.5008 (50.08%)")
print("   Paper Dir Acc excess returns:          64.52%")
print("   Paper Dir Acc volatility:              80.91%")

# ============================================================================
# 7. ALMACENAR OBJETOS PARA SIGUIENTE ETAPA
# ============================================================================

training_results_faithful = {
    'final_results': final_results,
    'methodology': 'expanding_window_faithful_to_paper',
    'periods': {
        'training': f"{training_start.strftime('%Y-%m')} to {training_end.strftime('%Y-%m')}",
        'validation': f"{validation_start.strftime('%Y-%m')} to {validation_end.strftime('%Y-%m')}",
        'test': f"{test_start.strftime('%Y-%m')} to {test_end.strftime('%Y-%m')}"
    },
    'scalers': {
        'feature_scaler': feature_scaler,
        'target_scalers': target_scalers
    },
    'metadata': {
        'feature_variables': feature_variables,
        'target_variables': target_variables,
        'total_test_predictions': len(test_mask.sum())
    }
}

print(f"\nüì¶ OBJETOS PARA ETAPA 5:")
print("   ‚úÖ Resultados finales por modelo y target")
print("   ‚úÖ Mejores hiperpar√°metros seleccionados")
print("   ‚úÖ Predicciones out-of-sample completas")
print("   ‚úÖ Metodolog√≠a fiel a Pinelis & Ruppert")

print("\nüéØ METODOLOG√çA IMPLEMENTADA:")
print("   ‚úÖ Expanding window validation")
print("   ‚úÖ Modelos separados por target")
print("   ‚úÖ Per√≠odos temporales correctos")
print("   ‚úÖ R¬≤ out-of-sample seg√∫n paper")
print("   ‚úÖ Evaluaci√≥n mes a mes en test set")

print("\n" + "="*70)
```
